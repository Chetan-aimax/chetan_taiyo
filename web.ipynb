```python
import bs4
import urllib.request
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
import csv
filename = 'music.csv'
f = open(filename,'w',newline = '')
music = csv.writer(f)
html = urlopen('https://dot.ca.gov/programs/procurement-and-contracts/contracts-out-for-bid')
bsobj = soup(html.read())
tbody = bsobj('table',{'class':'table'})[0].findAll('tr')
xl = []
for row in tbody:
    cols = row.findChildren(recursive = False)
    cols = [element.text.strip() for element in cols]
    music.writerow(cols) #Writing to CSV
    xl.append(cols)
import pandas as pd
df = pd.DataFrame(data = xl[1:],columns = xl[0])
```

```python
df.to_csv("Contracts.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# reading the database
data = pd.read_csv("Contracts.csv")

# Scatter plot with day against tip
plt.bar(data['End Date'], data['Event ID'])
plt.rcParams["figure.figsize"] = (2,5)
# Adding Title to the Plot
plt.title("bar chart")

# Setting the X and Y labels
plt.xlabel('End date')
plt.ylabel('Event id')

plt.show()
```
